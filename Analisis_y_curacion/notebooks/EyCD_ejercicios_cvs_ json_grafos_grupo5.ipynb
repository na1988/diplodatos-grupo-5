# Ejercicio 1: EyCD
Hacer Grafico a partir de tabla CSV

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
import sys
in_colab = 'google.colab' in sys.modules

if in_colab:
  BASE_DIR = "https://github.com/DiploDatos/AnalisisYCuracion/raw/master/"
else:
  BASE_DIR = ".."
```

```python
dataset = pd.read_csv(BASE_DIR + "/input/wtageinf.csv")
dataset.head(10)
```

```python
boy=dataset[dataset.Sex == 1]
girl=dataset[dataset.Sex == 2]
```

```python
boy.head()
```

```python
plt.figure(figsize=(10,6))
for column in boy.columns:
    if column[0] == "P":
            seaborn.set(font_scale=1.5)
            seaborn.lineplot(data=boy, x="Agemos", y=boy[column])
            plt.ylabel('Altura')
            plt.legend(["P3", "P5", "P10", "P25", "P50", "P75", "P90", "P95", "P97"], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
            plt.xlabel('Edad')
            plt.title('Percentil del crecimiento en altura en Niños')
```

```python
plt.figure(figsize=(10,6))
for column in girl.columns:
    if column[0] == "P":
            seaborn.set(font_scale=1.5)
            seaborn.lineplot(data=girl, x="Agemos", y=girl[column])
            plt.ylabel('Altura')
            plt.legend(["P3", "P5", "P10", "P25", "P50", "P75", "P90", "P95", "P97"], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
            plt.xlabel('Edad')
            plt.title('Percentil del crecimiento en altura en Niñas')
```
